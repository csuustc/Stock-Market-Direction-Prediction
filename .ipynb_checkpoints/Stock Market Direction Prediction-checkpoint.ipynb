# I. Introduction

It's a popular subject to predict the stock direction according to the history return and traded volume. In this project, we'll try different models such as logistic regression, linear discriminant analysis, quadratic discriminant analysis, and K-Nearest Neighbors to see which model get best performace.

The dataset we'll use is "Smarket" dataset, which consists of percentage returns for the S&P 500 stock index over 1250 days, from the beginning of 2001 until the end of 2005. For each date, we have recorded the percentage returns for each of the five previous trading days, $Lag1$ through $Lag5$ . We have also recorded $Volume$ (the number of shares traded on the previous day, in billions), $Today$ (the percentage return on the date in question) and $Direction$ (whether the market was Up or Down on this date).

# III. Model fitting and Prediction

## 3.1 Logistic Regresstion

Next, we will fit a logistic regression model in order to predict $Direction$ using $Lag1$ through $Lag5$ and $Volume$.

```python
import statsmodels.api as sm
X_sm = smarket.iloc[:, 1:7]
Y_sm = pd.get_dummies(smarket.Direction)[['Up']]

glm_fit_sm = sm.Logit(Y_sm, sm.add_constant(X_sm)).fit()
glm_fit_sm.summary()
```

The smallest p-value here is associated with $Lag1$. The negative coefficient for this predictor suggests that if the market had a positive return yesterday, then it is less likely to go up today. However, at a value of 0.15, the p-value is still relatively large, and so there is no clear evidence of a real association between $Lag1$ and $Direction$.

```python
glm_fit_sm.predict()
```

```python
glm_fit_sm.pred_table()
```

```python
print('Accuracy rate: ', (145+507)/1250)
```

The diagonal elements of the confusion matrix indicate correct predictions, while the off-diagonals represent incorrect predictions. Hence our model correctly predicted that the market would go up on 507 days and that it would go down on 145 days, for a total of 507 + 145 = 652 correct
predictions. So that the accuracy rate is 52.16%

But the true up ratio is 51.84%, which means if we predict all days up, the accuracy rate is still 51.84%, so it's not a quite acceptable model.

Considering the training error rate is often overly optimisticâ€”it tends to underestimate the test error rate. In order to better assess the accuracy of the logistic regression model in this setting, we can fit the model using part of the data, and then examine how well it predicts the held out
data. Now the whole dataset is splitted as training set, which contains observations from 2001 to 2004, and test set, which contains observations in 2005.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

train = smarket[smarket.Year < 2005]
test = smarket[smarket.Year >= 2005]

X_sk_train = train.iloc[:, 1:7]
Y_sk_train = train.Direction

X_sk_test = test.iloc[:, 1:7]
Y_sk_test = test.Direction

glm_model_t = LogisticRegression(fit_intercept=True, C=1e9)
glm_fit_t = glm_model_t.fit(X_sk_train, Y_sk_train)

conf_m = pd.DataFrame(confusion_matrix(Y_sk_train, glm_fit_t.predict(X_sk_train)))
conf_m.columns.name = 'Predicted'
conf_m.index.name = 'True'
conf_m
```

```python
accuracy_score(Y_sk_train, glm_fit_t.predict(X_sk_train))
```

Then use training set to fit logistic regression model. The result shows the accuracy rate of training set is 52.7%, which is lightly higher than the model fitted upon whole dataset.

Then, let's try to fit the model using test set.

```python
conf_m = pd.DataFrame(confusion_matrix(Y_sk_test, glm_fit_t.predict(X_sk_test)))
conf_m.columns.name = 'Predicted'
conf_m.index.name = 'True'
conf_m
```

```python
accuracy_score(Y_sk_test, glm_fit_t.predict(X_sk_test))
```

The results are rather disappointing: the test error rate is 52%, which is worse than random guessing! Of course this result is not all that surprising, given that one would not generally expect to be able to use previous daysâ€™ returns to predict future market performance.

Since most of the predictors are not significant in previous hypothesis testing, the logistic regression model had very underwhelming p-values associated with all of the predictors, and that the smallest p-value, though not very small, corresponded to $Lag1$. Perhaps by removing the variables that appear not to be helpful in predicting Direction, we can obtain a more effective model. Below we have refit the logistic regression using just $Lag1$ and $Lag2$, which seemed to have the highest predictive power in
the original logistic regression model.

```python
X_sk_train_s = train[['Lag1', 'Lag2']]
Y_sk_train_s = train.Direction

X_sk_test_s = test[['Lag1', 'Lag2']]
Y_sk_test_s = test.Direction

glm_fit_t_s = glm_model_t.fit(X_sk_train_s, Y_sk_train_s)

conf_m = pd.DataFrame(confusion_matrix(Y_sk_test_s, glm_fit_t_s.predict(X_sk_test_s)))
conf_m.columns.name = 'Predicted'
conf_m.index.name = 'True'
conf_m
```

```python
accuracy_score(Y_sk_test_s, glm_fit_t_s.predict(X_sk_test_s))
```

Now the results appear to be a little better: 56% of the daily movements have been correctly predicted. It is worth noting that in this case, a much simpler strategy of predicting that the market will increase every day will also be correct 56% of the time! This suggests a possible trading strategy of buying on days when the model predicts an increasing market, and avoiding trades on days when a decrease is predicted. Of course one would need to investigate more carefully whether this small improvement was real or just due to random chance.

Now we can try to predict the stock direction using $Lag1$ and $Lag2$. For example, if the return rate is 1.5% yesterday and -0.8% the 2 days before, then the direction will be:

```python
glm_fit_t_s.predict([[1.5, -0.8]])
```

## 3.2 Linear Discriminant Analysis

Now we'll perform LDA on the training set.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

lda_fit = LDA().fit(X_sk_train_s, Y_sk_train_s)
lda_fit.priors_
```

The LDA output indicates that $\hat{\pi{1}}$ = 0.492, $\hat{\pi{2}}$ = 0.508, which means in the training set, the observations going up are 50.8%

```python
lda_fit.means_
```

```python
lda_fit.scalings_
```

The coefficients of linear discriminants output provides the linear combination of $Lag1$ and $Lag2$ that are used to form the LDA decision rule. The following plots show the linear discriminants in different groups.

```python
ld_sc = X_sk_train_s.Lag1*lda_fit.scalings_[0] + X_sk_train_s.Lag2*lda_fit.scalings_[1]
ld = pd.DataFrame({'group': Y_sk_train_s, 'ld': ld_sc})

p = sns.FacetGrid(ld, col='group')
p.map(plt.hist, 'ld')
```

Then we'll fit the model using testing set.

```python
pd.DataFrame({'prob_0': lda_fit.predict_proba(X_sk_test_s)[:, 0], 
              'prob_1': lda_fit.predict_proba(X_sk_test_s)[:, 1], 
              'pred_class': lda_fit.predict(X_sk_test_s)}, 
             columns=['prob_0', 'prob_1', 'pred_class'])
```

```python
conf_m = pd.DataFrame(confusion_matrix(Y_sk_test_s, lda_fit.predict(X_sk_test_s)))
conf_m.columns.name = 'Predicted'
conf_m.index.name = 'True'
conf_m
```

```python
accuracy_score(Y_sk_test_s, lda_fit.predict(X_sk_test_s))
```

Here we can see, the result of LDA is almost same as logistic regression model, the accuracy rates of two models are both 56%

## 3.3 Quadratic Discriminant Analysis

We'll now fit a QDA model to the training set.

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA

qda_fit = QDA().fit(X_sk_train_s, Y_sk_train_s)
qda_fit.priors_
```

```python
qda_fit.means_
```

```python
qda_fit.scalings_
```

```python
pd.DataFrame({'prob_0': qda_fit.predict_proba(X_sk_test_s)[:, 0], 
              'prob_1': qda_fit.predict_proba(X_sk_test_s)[:, 1], 
              'pred_class': qda_fit.predict(X_sk_test_s)}, 
             columns=['prob_0', 'prob_1', 'pred_class'])
```

```python
conf_m = pd.DataFrame(confusion_matrix(Y_sk_test_s, qda_fit.predict(X_sk_test_s)))
conf_m.columns.name = 'Predicted'
conf_m.index.name = 'True'
conf_m
```

```python
accuracy_score(Y_sk_test_s, qda_fit.predict(X_sk_test_s))
```

Interestingly, the QDA predictions are accurate almost 60% of the time, even though the 2005 data was not used to fit the model. This level of accuracy is quite impressive for stock market data, which is known to be quite hard to model accurately. This suggests that the quadratic form assumed by QDA may capture the true relationship more accurately than the linear forms assumed by LDA and logistic regression. However, we recommend evaluating this methodâ€™s performance on a larger test set before betting
that this approach will consistently beat the market!

## 3.4 K-Nearest Neighbors

At last, we'll try to fit K-Nearest Neighbors and see its performance. Let's start from K = 1.

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model_1 = KNeighborsClassifier(n_neighbors=1)
knn_fit = knn_model_1.fit(X_sk_train_s, Y_sk_train_s)
pd.DataFrame({'prob_0': knn_fit.predict_proba(X_sk_train_s)[:, 0], 
              'prob_1': knn_fit.predict_proba(X_sk_train_s)[:, 1], 
              'pred_class': knn_fit.predict(X_sk_train_s)}, 
             columns=['prob_0', 'prob_1', 'pred_class'])
```

```python
conf_m = pd.DataFrame(confusion_matrix(Y_sk_test_s, knn_fit.predict(X_sk_test_s)))
conf_m.columns.name = 'Predicted'
conf_m.index.name = 'True'
conf_m
```

```python
accuracy_score(Y_sk_test_s, knn_fit.predict(X_sk_test_s))
```

The accuracy rate is only 0.5. Acutually K = 1 can't fit well under most circumstance. Let's try K = 3.

```python
knn_model_3 = KNeighborsClassifier(n_neighbors=3)
knn_fit_3 = knn_model_3.fit(X_sk_train_s, Y_sk_train_s)

conf_m = pd.DataFrame(confusion_matrix(Y_sk_test_s, knn_fit_3.predict(X_sk_test_s)))
conf_m.columns.name = 'Predicted'
conf_m.index.name = 'True'
conf_m
```

```python
accuracy_score(Y_sk_test_s, knn_fit_3.predict(X_sk_test_s))
```

The accuracy rate of K=3 model is 53.17, which is pretty better than K=1. But it's still little improvement than random guessing.
